Next: unite the different year dataframes into one.  

```python
frames = [df_energy_api_2018,df_energy_api_2017, df_energy_api_2016, df_energy_api_2015]
```

```python
df_energy_api_2015_to_2018 = pd.concat(frames)
```

```python
df_energy_api_2015_to_2018.sample(3)
```

Eliminate the column "values" as it does not contribute to testing our hypothesis. 

```python
df_energy_api_2015_to_2018 = df_energy_api_2015_to_2018[["percentage", "datetime"]]
```

```python
df_energy_api_2015_to_2018.sample(3)
```

Renaming: 
- "percentage" to : "renewable energy generation percentage"
- "datetime" to "date"

```python
df_energy_api_2015_to_2018 = df_energy_api_2015_to_2018.rename(columns={'percentage': 'renewable energy generation percentage'})
```

```python
df_energy_api_2015_to_2018 = df_energy_api_2015_to_2018.rename(columns={'datetime': 'date'})
```

Add a new column for : "non-renewable energy generation percentage". 

```python
df_energy_api_2015_to_2018["non-renewable energy generation percentage"] = (1 - df_energy_api_2015_to_2018["renewable energy generation percentage"])
```

```python
df_energy_api_2015_to_2018.sample(3)
```

Re-organizing the columns order

```python
df_energy_api_2015_to_2018 = df_energy_api_2015_to_2018[["renewable energy generation percentage", "non-renewable energy generation percentage", "date"]]
```

```python
df_energy_api_2015_to_2018.sample(3)
```

Elimating rows to make even with other dataset: 

```python
df_energy_api_2015_to_2018 = df_energy_api_2015_to_2018.sort_values('date')
```

```python
df_energy_api_2015_to_2018 = df_energy_api_2015_to_2018.drop(df_energy_api_2015_to_2018.index[[0]])
```

```python
df_energy_api_2015_to_2018.tail(3)
```

Change the column and only keep, year-month-day, deleating hours

```python
df_energy_api_2015_to_2018.info()
```

```python
df_energy_api_2015_to_2018['date'] = pd.to_datetime(df_energy_api_2015_to_2018['date']).dt.date
```

```python
df_energy_api_2015_to_2018.head(3)
```

Final dataframe:

```python
df_energy_api_2015_to_2018_final = df_energy_api_2015_to_2018
```

```python
df_energy_api_2015_to_2018_final.tail(3)
```
